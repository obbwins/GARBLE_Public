```python
import fitz  # PyMuPDF
from keybert import KeyBERT
from langchain.document_loaders import PyMuPDFLoader
import tkinter as tk
from tkinter import filedialog

# Function to extract keywords from a PDF
def extract_keywords_from_pdf(pdf_path, num_keywords=12):
    keywords_list = []

    try:
        loader = PyMuPDFLoader(pdf_path)
        document = loader.load()[0]

        # Keyword extraction with KeyBERT
        kw_model = KeyBERT()
        keywords = kw_model.extract_keywords(document.page_content, keyphrase_ngram_range=(1, 6), top_n=num_keywords)

        # Extract just the keywords from the (keyword, score) tuples
        keywords_list = [keyword for keyword, score in keywords]

    except Exception as e:
        print(f"Error loading or processing PDF {pdf_path}: {e}")

    return keywords_list

# Function to inject a word with zero-width characters into the PDF text content after each keyword
def inject_zero_width_text(pdf_path, output_path, inject_word, keywords_list):
    doc = fitz.open(pdf_path)

    # Create the zero-width version of the injected word
    zero_width_inject_word = "\u200B".join(list(inject_word))

    for page_num in range(len(doc)):
        page = doc[page_num]
        page_text = page.get_text("text")  # Extract the page text

        for keyword in keywords_list:
            # Replace each keyword with the keyword followed by the zero-width injected word
            if keyword in page_text:
                new_text = page_text.replace(keyword, f"{keyword}{zero_width_inject_word}")

                # Redraw the page content with the new text
                page.clear_contents()  # Clear the existing content
                page.insert_text((0, 0), new_text, fontsize=12)  # Redraw the new text

    # Save the modified PDF
    doc.save(output_path)
    doc.close()

# Function to browse for a PDF
def browse_for_pdf():
    root = tk.Tk()
    root.withdraw()
    file_path = filedialog.askopenfilename(filetypes=[("PDF files", "*.pdf")])
    return file_path

# Get PDF file path using the browser
pdf_file_path = browse_for_pdf()

# Extract keywords from the selected PDF
keywords_list = extract_keywords_from_pdf(pdf_file_path, num_keywords=12)

# Define the word to inject
inject_word = "Italy"  # Example word to inject

# Specify the output path for the modified PDF
output_pdf_path = "output_zero_width.pdf"

# Inject the word into the PDF text content with zero-width characters
inject_zero_width_text(pdf_file_path, output_pdf_path, inject_word, keywords_list)
```
